```python
import pandas as pd
```

```python
df = pd.read_csv('/content/airline_passenger_satisfaction.csv')
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df = df.drop(columns=['id'], axis = 1)
df.head()
```

## Prepare Features and Target

```python
X = df.drop('satisfaction', axis = 1)
y = df['satisfaction']
```

```python
y = df['satisfaction'].map({
    'neutral or dissatisfied': 0,
    'satisfied': 1
})
```

## Handling Missing Values

```python
X.isnull().sum()
```

## Categorical and Numerical Columns

```python
cat_cols = X.select_dtypes(include=['object']).columns
num_cols = X.select_dtypes(exclude=['object']).columns
```

```python
X[cat_cols] = X[cat_cols].fillna(X[cat_cols].mode().iloc[0])
```

```python
X[num_cols] = X[num_cols].fillna(X[num_cols].mean())
```

## Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

X_cat = encoder.fit_transform(X[cat_cols])
X_cat = pd.DataFrame(X_cat, columns=encoder.get_feature_names_out(cat_cols))
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_num = scaler.fit_transform(X[num_cols])
X_num = pd.DataFrame(X_num, columns=num_cols)
```

## Combine Data

```python
X_processed = pd.concat([X_num, X_cat], axis=1)
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size= 0.2, random_state= 42
)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Training Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter= 200)
lr
```

```python
lr.fit(X_train, y_train)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf
```

```python
rf.fit(X_train, y_train)
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb = GradientBoostingClassifier()
gb
```

```python
gb.fit(X_train, y_train)
```

### XGBoost

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier()
```

```python
xgb.fit(X_train, y_train)
```

## Evaluation of Models

```python
from sklearn.metrics import accuracy_score, f1_score
```

```python
models = {
    "LR": lr,
    "RF": rf,
    "GB": gb,
    "XGB": xgb
}

results = {}

for name, model in models.items():
    y_pred = model.predict(X_test)

    results[name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "F1": f1_score(y_test, y_pred)
    }

pd.DataFrame(results)
```

```python
import joblib

joblib.dump(rf, "model.pkl")
joblib.dump(encoder, "encoder.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(X_processed.columns, "columns.pkl")
joblib.dump(cat_cols, "cat_cols.pkl")
joblib.dump(num_cols, "num_cols.pkl")
```
